```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import kagglehub
import os

#Extract
# Download latest version of dataset off of Kaggle
path = kagglehub.dataset_download("emirhanai/social-media-usage-and-emotional-well-being")

# Load CSV files (train, test and val)
train_path = os.path.join(path, "train.csv")
test_path = os.path.join(path, "test.csv")
val_path = os.path.join(path, "val.csv")

print("Dataset files extracted:", train_path, test_path, val_path)

# Load in DataFrames
train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)
val_df = pd.read_csv(val_path)

# Combine all data for analysis using concat function
df = pd.concat([train_df, test_df, val_df], ignore_index=True)

# Transform step
def clean_data(df):
    """Cleans and preprocesses the dataset."""
    
    # Drop duplicates
    df = df.drop_duplicates()
    
    # Handle missing values
    df = df.fillna(df.median(numeric_only=True))  # Fill missing numeric values with median
    df = df.fillna(df.mode().iloc[0])  # Fill categorical missing values with mode

    # Identify rows where Age and Gender are swapped
    swap_mask = df["Age"].isin(["Male", "Female", "Non-binary"])

    # swap Age and Gender values where necessary
    df.loc[swap_mask, ["Age", "Gender"]] = df.loc[swap_mask, ["Gender", "Age"]].values
    
    # Convert categorical columns to lowercase
    df["Gender"] = df["Gender"].str.lower()
    df["Platform"] = df["Platform"].str.lower()
    df["Dominant_Emotion"] = df["Dominant_Emotion"].str.lower()

    #Ensures age is numeric
    df["Age"] = pd.to_numeric(df["Age"], errors="coerce")

    #Create list of valid genders so that other answers (Like Marie) are cleaned out
    valid_genders = ["male", "female", "non-binary"]
    df = df[df["Gender"].isin(valid_genders)]

    # Drop null values for age
    df = df.dropna(subset=["Age"])  
    return df

df = clean_data(df)

# Save cleaned data to a new dataset
df.to_csv("cleaned_social_media_data.csv", index=False)
print("Cleaned dataset saved as cleaned_social_media_data.csv")

#EDA
def perform_eda(df):
    """Performs exploratory data analysis on the dataset."""
    
    print("\n Summary Statistics: ")
    print(df.describe())

    print("\n Gender Distribution: ")
    print(df["Gender"].value_counts())

    print("\n Platform Usage: ")
    print(df["Platform"].value_counts())

    print("\n Most Common Emotional States: ")
    print(df["Dominant_Emotion"].value_counts())

    # Visualization settings
    sns.set_style("whitegrid")

    # Distribution of Daily Usage Time
    plt.figure(figsize=(8, 4))
    sns.histplot(df["Daily_Usage_Time (minutes)"], bins=30, kde=True, color="blue")
    plt.title("Distribution of Daily Usage Time")
    plt.xlabel("Minutes")
    plt.ylabel("Frequency")
    plt.show()

    # Daily Usage Time per Platform Boxplot
    plt.figure(figsize=(10, 5))
    sns.boxplot(x="Platform", y="Daily_Usage_Time (minutes)", data=df)
    plt.xticks(rotation=45)
    plt.title("Daily Usage Time per Platform")
    plt.show()

    # Emotional State Distribution
    plt.figure(figsize=(8, 4))
    sns.countplot(x=df["Dominant_Emotion"], palette="coolwarm")
    plt.xticks(rotation=45)
    plt.title("Distribution of Dominant Emotions")
    plt.xlabel("Emotion")
    plt.ylabel("Count")
    plt.show()

    # Correlation Heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
    plt.title("Feature Correlation Heatmap")
    plt.show()

    #Gender vs. Daily Usage Time Boxplot
    plt.figure(figsize=(8, 5))
    sns.boxplot(x="Gender", y="Daily_Usage_Time (minutes)", data=df, palette="pastel")
    plt.title("Daily Usage Time by Gender")
    plt.xlabel("Gender")
    plt.ylabel("Daily Usage Time (minutes)")
    plt.show()

    #Gender vs. Daily Usage Time Violin Plot
    plt.figure(figsize=(8, 5))
    sns.violinplot(x="Gender", y="Daily_Usage_Time (minutes)", data=df, palette="muted", inner="quartile")
    plt.title("Daily Usage Time by Gender")
    plt.xlabel("Gender")
    plt.ylabel("Daily Usage Time (minutes)")
    plt.show()

    # Age Distribution by Gender histogram
    plt.figure(figsize=(8, 5))
    sns.histplot(x="Age", hue="Gender", data=df, element="step", kde=True, palette="husl", bins=30)
    plt.title("Age Distribution by Gender")
    plt.xlabel("Age")
    plt.ylabel("Count")
    plt.show()

    # Daily Usage Time Trend by Age regression
    plt.figure(figsize=(10, 5))
    sns.scatterplot(x="Age", y="Daily_Usage_Time (minutes)", data=df, alpha=0.6)
    sns.regplot(x="Age", y="Daily_Usage_Time (minutes)", data=df, scatter=False, color="red")
    plt.title("Daily Usage Time vs. Age")
    plt.xlabel("Age")
    plt.ylabel("Daily Usage Time (minutes)")
    plt.show()

    # Emotional State Distribution by Gender count plot
    plt.figure(figsize=(10, 5))
    sns.countplot(x="Dominant_Emotion", hue="Gender", data=df, palette="coolwarm")
    plt.xticks(rotation=45)
    plt.title("Emotional State Distribution by Gender")
    plt.xlabel("Emotion")
    plt.ylabel("Count")
    plt.legend(title="Gender")
    plt.show()
    
perform_eda(df)
```
